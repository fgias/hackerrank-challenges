```python
#!/bin/python3

import math
import os
import random
import re
import sys



#
# Complete the 'getIdealNums' function below.
#
# The function is expected to return a LONG_INTEGER.
# The function accepts following parameters:
#  1. LONG_INTEGER low
#  2. LONG_INTEGER high
#

def isIdeal(num):
    while (num%3 == 0) or (num%5 == 0):
        if num%3 == 0:
            num = num/3
        else:
            num = num/5

    if num == 1:
        return True
    else:
        return False 

def getIdealNums(low, high):
    lowest_found = False
    lowest = low

    highest_found = False 
    highest = high

    while lowest_found == False:
        if isIdeal(lowest):
            lowest_found == True
        else: 
            lowest += 1

    while highest_found == False:
        if isIdeal(highest):
            highest_found == True
        else: 
            highest += -1

    list = []

    ideal = lowest * 3
    while ideal < highest:
        list.append(ideal)
        ideal = 3 * ideal

    ideal = lowest * 5
    while ideal < highest:
        list.append(ideal)
        ideal = 5 * ideal
    
    list.append(lowest)
    list.append(highest)

    return len(list)
```
