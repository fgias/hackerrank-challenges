## Calculating Maximum Drawdown

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

We are given a list of daily gains and losses `L`.

```python
L = [0.1, 0.1, 0.02, -0.2, 0.1, -.5, -.8, -1.0, 1.3, -.9, .5, .5, .5, .5, .5, .5, -.6, -.6, .6, .8, .9, 1, 2]
```

```python
sns.set_style('whitegrid')
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(len(L)),L, marker='o')
plt.xlabel('time')
plt.title('Daily Gains/Losses')
```

Calculate the cumulative gain/loss.

```python
current = 0 
M = []
for i in range(len(L)):
    M.append(current)
    current += L[i]
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(len(M)),M, marker='o')
plt.xlabel('time')
plt.title('Cumulative Gain/Loss')
```

Calculate the running maximum (last highest point).

```python
curr_high = 0
last_high = []
for i in range(len(M)):
    if M[i] > curr_high:
        curr_high = M[i]
        last_high.append(curr_high)
    else:
        last_high.append(curr_high)
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(len(last_high)),last_high, marker='o')
plt.xlabel('time')
plt.title('Running Max')
```

Calculate the difference between the last high and each point after it.

```python
drawdown = [x - y for x,y in zip(M, last_high)]
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(len(drawdown)),drawdown, marker='o')
plt.xlabel('time')
plt.title('Drawdown')
```

Hence, calculate the maximum drawdown.

```python
min(drawdown)
```

Putting it all in a function.

```python
def maxDrawdown(L):
    current = 0 
    M = []
    for i in range(len(L)):
        M.append(current)
        current += L[i]

    curr_high = 0
    last_high = []
    for i in range(len(M)):
        if M[i] > curr_high:
            curr_high = M[i]
            last_high.append(curr_high)
        else:
            last_high.append(curr_high)
    
    drawdown = [x - y for x,y in zip(M, last_high)]

    return min(drawdown)
```

```python
maxDrawdown(L)
```
